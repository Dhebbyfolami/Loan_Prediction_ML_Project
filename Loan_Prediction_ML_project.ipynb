```python
#import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read the csv file
data = pd.read_csv("trainLoan.csv")
```

```python
#let's view the first five rows
data.head()
```

```python
#let's check the info of all the data to see if there is any null value
data.info()
```

```python
#lets check for the null value
data.isnull().sum()
```

```python
#lets check for columns that are non-numerica
obj = (data.dtypes == "object")
print("Categorical variables: ", len(list(obj[obj].index)))
```

```python
#let's drop the id to keep only useful features. Because the ID is just the unique identier not to predict the loan
data.drop(["Loan_ID"], axis = 1, inplace = True)
```

```python
data.head(5)
```

```python
#lets plot a bar chart for the column non-numerica list
obj = (data.dtypes == "object")
object_cols = list(obj[obj].index)
plt.figure(figsize=(10, 34))
index = 1

for col in object_cols:
    y = data[col].value_counts()
    plt.subplot(12, 4, index)
    plt.xticks(rotation = 90)
    sns.barplot(x = list(y.index), y=y)
    index += 3
```

```python
#convert text to number using LabelEncoding
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
obj = (data.dtypes == "object")
for col in list(obj[obj].index):
    data[col] = label_encoder.fit_transform(data[col])
```

```python
data.head(5)
```

```python
#check for the remaining category columns
obj = (data.dtypes == "object")
print('Categorical variables: ', len(list(obj[obj].index)))
```

```python
#Lets use hitmap to check for probability of higher earner for higher loan amount. And also to show features that affect the loan status
plt.figure(figsize=(12, 6))
sns.heatmap(data.corr(), cmap = "BrBG", fmt=".2f", linewidths = 2, annot = True)
```

```python
#Lets use gender and marrital status to show if the married men or women will have high probability of getting the loan. 
#the loan status will be use to determine whether loan is approved or not with the use of colors to the bar
sns.catplot(x = "Gender", y = "Married", hue = "Loan_Status", kind = "bar", data = data)
```

```python
#lets fill in the missing data or values(e.g NaN)
for col in data.columns:
    data[col] = data[col].fillna(data[col].mean())

data.isna().sum()
```

```python
data.head(5)
```

```python
from sklearn.model_selection import train_test_split

X = data.drop(["Loan_Status"], axis = 1)
Y = data["Loan_Status"]
X.shape, Y.shape

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.4, random_state = 1)
```

```python
#lets train 4 different models Random Forest, KNeighbors, Support Vector Machine and Logistic Regression to test how well they perform
#and print the accuracy for each. this help us  compare which model learn best
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

from sklearn import metrics

knn = KNeighborsClassifier(n_neighbors = 3)
rfc = RandomForestClassifier(n_estimators = 7, criterion = "entropy", random_state = 7)
svc = SVC()
lc = LogisticRegression()

for clf in (rfc, knn, svc, lc):
    clf.fit(X_train, Y_train)
    Y_pred = clf.predict(X_train)
    print("Accuracy score of", clf.__class__.__name__, "=", 100*metrics.accuracy_score(Y_train, Y_pred))
```

```python
#To see how good each model really is at predicting let's try to test it again
for clf in (rfc, knn, svc, lc):
    clf.fit(X_train, Y_train)
    Y_pred = clf.predict(X_train)
    print("Accuracy score of", clf.__class__.__name__, "=", 100*metrics.accuracy_score(Y_train, Y_pred))
```

```python
#In the result above it shows the models are good. RandomForestClassifier and LogisticRegression gives the more perfect result.
```
